```python
# Numerical and Categorical features

num_cols = df.select_dtypes(include=np.number).columns
print(num_cols)
```

```python
categorical_cols = df.select_dtypes(include="str").columns

print(categorical_cols)
```

```python
# Distribution of target variable

plt.figure(figsize=(10,6))
sns.histplot(df["price_usd"], bins=30, kde=True)
plt.title("Distribution of EV Prices")
plt.show()
```

```python
# Correlation Heatmap

plt.figure(figsize=(16,10))
sns.heatmap(df[num_cols].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Brand and Price

plt.figure(figsize=(14,6))
sns.boxplot(x="brand", y="price_usd", data=df)
plt.xticks(rotation=45)
plt.title("Brand vs Price")
plt.show()
```

```python
# Battery capacity and range

plt.figure(figsize=(10,6))
sns.scatterplot(x="battery_capacity_kwh", y="range_miles", data=df)
plt.title("Battery Capacity vs Range")
plt.show()
```

```python
# Sales analysis

plt.figure(figsize=(12,6))
sns.barplot(x="market_segment", y="annual_sales_units", data=df)
plt.title("Market Segment vs Annual Sales")
plt.show()
```

```python
# Top 10 most expensive EVs

print(df[["brand", "model", 'price_usd']].sort_values(by="price_usd", ascending=False).head(10))
```
